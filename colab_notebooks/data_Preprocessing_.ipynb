```python
import numpy as np
import pandas as pd
from google.colab import files
import io
import os
from google.colab import drive

# To import datasets from Kaggle
# 1. Follow this tutorial (https://www.geeksforgeeks.org/how-to-import-kaggle-datasets-directly-into-google-colab/)
# 2. OR Download the dataset to your Google drive and follow this code (https://www.kaggle.com/datasets/rajugc/imdb-movies-dataset-based-on-genre/data)

drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/movies'


action_movie = pd.read_csv(f'{file_path}/action.csv')
crime_movie = pd.read_csv(f'{file_path}/crime.csv')
adventure_movie = pd.read_csv(f'{file_path}/adventure.csv')
thriller_movie = pd.read_csv(f'{file_path}/thriller.csv')
family_movie = pd.read_csv(f'{file_path}/family.csv')
mystery_movie = pd.read_csv(f'{file_path}/mystery.csv')
scifi_movie = pd.read_csv(f'{file_path}/scifi.csv')
history_movie = pd.read_csv(f'{file_path}/history.csv')
sports_movie = pd.read_csv(f'{file_path}/sports.csv')
animation_movie = pd.read_csv(f'{file_path}/animation.csv')
biography_movie = pd.read_csv(f'{file_path}/biography.csv')
horror_movie = pd.read_csv(f'{file_path}/horror.csv')
fantasy_movie = pd.read_csv(f'{file_path}/fantasy.csv')
romance_movie = pd.read_csv(f'{file_path}/romance.csv')
```

```python
action_movie['genre_1'] = 'Action'
crime_movie['genre_1'] = 'Crime'
adventure_movie['genre_1'] = 'Adventure'
thriller_movie['genre_1'] = 'Thriller'
family_movie['genre_1'] = 'Family'
mystery_movie['genre_1'] = 'Mystery'
scifi_movie['genre_1'] = 'Sci-Fi'
history_movie['genre_1'] = 'History'
sports_movie['genre_1'] = 'Sports'
animation_movie['genre_1'] = 'Animation'
biography_movie['genre_1'] = 'Biography'
horror_movie['genre_1'] = 'Horror'
fantasy_movie['genre_1'] = 'Fantasy'
romance_movie['genre_1'] = 'Romance'
```

```python
# Concatenate all dataframes
df = pd.concat([action_movie, crime_movie, adventure_movie, thriller_movie,
                family_movie, mystery_movie, scifi_movie, history_movie,
                sports_movie, animation_movie, biography_movie,
                horror_movie, fantasy_movie, romance_movie])
```

```python
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
df.head()
```

```python
df['genres'] = df['genre'].str.split(', ')
df = df.drop(columns=['star','genre','director','movie_id','genre_1','certificate','runtime','rating','director_id','star_id','votes','gross(in $)'])  # Delete a column Delete a column
df.head()
```

```python
# Count duplicates by both columns
duplicate_counts = df.groupby(['movie_name', 'year']).size()

# Filter to only titles that appear more than once
duplicates = duplicate_counts[duplicate_counts > 1]
df = df.drop_duplicates(subset=['movie_name', 'year'], keep='first')
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Flatten all genre lists and count occurrences
all_genres = [genre for sublist in df['genres'] for genre in sublist]
genre_counts = Counter(all_genres)
print(genre_counts)
print(len(genre_counts))
# western, film-noir, reality-tv, news, game-show, adult, talk-show, short, 9051.0

# Convert to DataFrame for plotting
genre_df = pd.DataFrame.from_dict(genre_counts, orient='index').reset_index()
genre_df.columns = ['Genre', 'Count']
genre_df = genre_df.sort_values('Count', ascending=False)

print(f"Unique genres: {len(genre_counts)}")
print(genre_df.head())
plt.figure(figsize=(12, 6))
plt.bar(genre_df['Genre'], genre_df['Count'], color='skyblue')
plt.title('Genre Frequency before data preprocessing')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
unwanted_genres = [
    'Western', 'Film-Noir', 'News',
    'Game-Show', 'Adult', 'Talk-Show', 'Short', '9051.0', 'Reality-TV', "Sport", "Musical", "Music", "War","Game-Show"
]
# Keep only rows where NONE of the unwanted genres appear
filtered_df = df[
    ~df['genres'].apply(lambda genres: any(genre in unwanted_genres for genre in genres))
]
remaining_genres = set(genre for sublist in filtered_df['genres'] for genre in sublist)
print("Remaining genres:", remaining_genres)
all_genres = [genre for sublist in filtered_df['genres'] for genre in sublist]
genre_counts = Counter(all_genres)
print(genre_counts)
print(len(genre_counts))
```

```python
# Flatten all genres to identify unique ones
all_genres = set(genre for sublist in filtered_df['genres'] for genre in sublist)
print(f"Unique genres: {len(all_genres)}")
```

```python
# Cleaning up Description Column

filtered_df['description'] = filtered_df['description'].str.replace('See full summary', '', regex=False)
filtered_df = filtered_df[filtered_df['description'].str.len() >= 50]
filtered_df['description'] = filtered_df['description'].str.replace('¬ª','', regex=True)
all_genres = [genre for sublist in filtered_df['genres'] for genre in sublist]
genre_counts = Counter(all_genres)
print(genre_counts)
print(len(genre_counts))
```

```python
# Get indices of movies that include "Drama"
drama_movies = filtered_df[filtered_df['genres'].apply(lambda x: 'Drama' in x)].copy()

# Randomly select 50,000 Drama movies to remove (or all if fewer exist)
n_to_remove = min(50000, len(drama_movies))
drama_to_remove = drama_movies.sample(n=n_to_remove, random_state=42)

# Remove these movies from the original DataFrame
df_filtered = filtered_df.drop(drama_to_remove.index)
```

```python
import matplotlib.pyplot as plt
# Flatten all genre lists and count occurrences
all_genres = [genre for sublist in df_filtered['genres'] for genre in sublist]
genre_counts = Counter(all_genres)
print(genre_counts)
print(len(genre_counts))
# Convert to DataFrame for plotting
genre_df = pd.DataFrame.from_dict(genre_counts, orient='index').reset_index()
genre_df.columns = ['Genre', 'Count']
genre_df = genre_df.sort_values('Count', ascending=False)

print(f"Unique genres: {len(genre_counts)}")
print(genre_df.head())
plt.figure(figsize=(12, 6))
plt.bar(genre_df['Genre'], genre_df['Count'], color='skyblue')
plt.title('Genre Frequency')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import ast
save_path = '/content/drive/MyDrive/movies/5000_movies.csv'
df = pd.read_csv(save_path)
df['genres'] = df['genres'].apply(ast.literal_eval)
```

```python
# prompt: remove genre Drama from the column genre list

def remove_drama(genre_list):
  if isinstance(genre_list, list):
    return [genre for genre in genre_list if genre != 'Drama']
  else:
    return genre_list

df['genres'] = df['genres'].apply(remove_drama)
```

```python
import matplotlib.pyplot as plt
# Flatten all genre lists and count occurrences
all_genres = [genre for sublist in df['genres'] for genre in sublist]
genre_counts = Counter(all_genres)
print(genre_counts)
print(len(genre_counts))
# Convert to DataFrame for plotting
genre_df = pd.DataFrame.from_dict(genre_counts, orient='index').reset_index()
genre_df.columns = ['Genre', 'Count']
genre_df = genre_df.sort_values('Count', ascending=False)

print(f"Unique genres: {len(genre_counts)}")
print(genre_df.head())
plt.figure(figsize=(12, 6))
plt.bar(genre_df['Genre'], genre_df['Count'], color='skyblue')
plt.title('Genre Frequency')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
